```python
heetch_df["Point"] = heetch_df.apply(lambda x: Point(x['Origin Latitude'], x['Origin Longitude']), axis=1)
```

```python
heetch_df.head()
```

```python
heetch_df.loc[heetch_df['Point'].apply(lambda x: x.within(R)) == True,'venue_type'] = 'Roissy'
heetch_df.loc[heetch_df['Point'].apply(lambda x: x.within(L)) == True,'venue_type'] = 'Gare de Lyon'
heetch_df.loc[heetch_df['Point'].apply(lambda x: x.within(O)) == True,'venue_type'] = 'Orly'
heetch_df.loc[heetch_df['Point'].apply(lambda x: x.within(A)) == True,'venue_type'] = "Gare d'Austerlitz"
heetch_df.loc[heetch_df['Point'].apply(lambda x: x.within(M)) == True,'venue_type'] = "Gare de Montparnasse"
heetch_df.loc[heetch_df['Point'].apply(lambda x: x.within(E)) == True,'venue_type'] = "Gare de L'Est"
heetch_df.loc[heetch_df['Point'].apply(lambda x: x.within(N)) == True,'venue_type'] = "Gare du Nord"
#heetch_df.loc[heetch_df['Point'].apply(lambda x: x.within(F)) == True,'venue_type'] = "France"
```
